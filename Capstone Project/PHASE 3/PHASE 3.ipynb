    districts = 'SELECT c.STATE/UT, c.YEAR, c.DISTRICT, c.MURDER
    FROM crime_data AS c
    INNER JOIN (
        SELECT YEAR, MAX(MURDER) AS max_murder
        FROM crime_data
        GROUP BY YEAR
        )
    AS subq ON c.YEAR = subq.YEAR AND c.MURDER = subq.max_murder
