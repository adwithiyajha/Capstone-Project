```python
df = df.drop(columns = ['STATE AND YEAR'], axis=1)
df
```

```python
df['LITERACY RATE'] = df['LITERACY RATE'].astype(float)
df.info()
```

Now, let's use iterative imputer on all feature columns with null values

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
ii = IterativeImputer()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
iter_col = pd.DataFrame(ii.fit_transform(df[['LITERACY RATE', 'AREA (in sq. km)','CRIMES AGAINST CHILDREN', 'JUVENILES APPREHENDED','POLICEMEN KILLED', 'THEFT CASES REPOERTED', 'THEFT VALUE TOTAL','TOTAL  Dacoity', 'TOTAL  Robbery', 'TOTAL  Burglary', 'TOTAL  Theft', 'Total Petty Crime', 'Human Rights Violation Cases', 'Policemen Chargesheeted', 'Polimen Charged']]),
                        columns = ['LITERACY RATE_ITER', 'AREA (in sq. km)_ITER','CRIMES AGAINST CHILDREN_ITER', 'JUVENILES APPREHENDED_ITER','POLICEMEN KILLED_ITER', 'THEFT CASES REPOERTED_ITER', 'THEFT VALUE TOTAL_ITER','TOTAL  Dacoity_ITER', 'TOTAL  Robbery_ITER', 'TOTAL  Burglary_ITER', 'TOTAL  Theft_ITER', 'Total Petty Crime_ITER', 'Human Rights Violation Cases_ITER', 'Policemen Chargesheeted_ITER', 'Polimen Charged_ITER'])

df = df.join(iter_col)
df
```

```python
drop_columns = ['LITERACY RATE', 'AREA (in sq. km)','CRIMES AGAINST CHILDREN', 'JUVENILES APPREHENDED','POLICEMEN KILLED', 'THEFT CASES REPOERTED', 'THEFT VALUE TOTAL','TOTAL  Dacoity', 'TOTAL  Robbery', 'TOTAL  Burglary', 'TOTAL  Theft', 'Total Petty Crime', 'Human Rights Violation Cases', 'Policemen Chargesheeted', 'Polimen Charged']
```

```python
df = df.drop(columns = drop_columns, axis=1)
df
```
